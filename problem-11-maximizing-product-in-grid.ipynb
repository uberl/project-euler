```python
AR = [list(map(int, line.split(" "))) for line in INPUT.split("\n")]
```

```python
import math
```

```python
N = 4  # Number of adjacent digits
maximum = 1

MAX_INDEX = len(AR) - N + 1

for i in range(MAX_INDEX):
    for j in range(MAX_INDEX):
        horizontal_slice = [AR[i][j + x] for x in range(N)]
        vertical_slice = [AR[i + x][j] for x in range(N)]
        diagonal_slice = [AR[i + x][j + x] for x in range(N)]
        reverse_diagonal_slice = [AR[i + x][j + (N - 1) - x] for x in range(N)] #top right to bottom left index offset = 3 !!

        for slice in [
            vertical_slice,
            horizontal_slice,
            diagonal_slice,
            reverse_diagonal_slice,
        ]:
            product = math.prod(slice)
            if product > maximum:
                maximum = product

maximum
```
